## Portabilidad num칠rica m칩vil: altas mensuales por operador
**Proporciona informaci칩n sobre las preferencias de los usuarios al elegir un nuevo operador y la competencia entre los diferentes operadores en el mercado de telefon칤a m칩vil**

```python
df = pd.read_csv('Sets/Movil/Portabilidad.csv')
df.head()
```

```python
# Llenamos los valores nulos con 0
df = df.fillna(0)
```

```python
df['A침o'] = df['A침o'].astype('category')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertimos las columnas relevantes a tipo num칠rico
df['Personal'] = df['Personal'].astype(float)
df['Claro'] = df['Claro'].astype(float)
df['Nextel'] = df['Nextel'].astype(float)
df['Movistar'] = df['Movistar'].astype(float)
df['Otros'] = df['Otros'].astype(float)

# Agrupamos por a침o y calcular el crecimiento promedio anual de las compa침칤as
crecimiento_personal = df.groupby('A침o')['Personal'].mean().pct_change() * 100
crecimiento_claro = df.groupby('A침o')['Claro'].mean().pct_change() * 100
crecimiento_nextel = df.groupby('A침o')['Nextel'].mean().pct_change() * 100
crecimiento_movistar = df.groupby('A침o')['Movistar'].mean().pct_change() * 100
crecimiento_otros = df.groupby('A침o')['Otros'].mean().pct_change() * 100

# Sacamos los a침os 칰nicos
a침os = df['A침o'].unique()

# Creamos la gr치fica de l칤neas para mostrar el crecimiento de las compa침칤as
plt.figure(figsize=(8, 6))
plt.plot(a침os[1:], crecimiento_personal[1:], label='Personal')
plt.plot(a침os[1:], crecimiento_claro[1:], label='Claro')
plt.plot(a침os[1:], crecimiento_nextel[1:], label='Nextel')
plt.plot(a침os[1:], crecimiento_movistar[1:], label='Movistar')
plt.plot(a침os[1:], crecimiento_otros[1:], label='Otros')
plt.xlabel('A침o')
plt.ylabel('Crecimiento porcentual')
plt.title('Crecimiento de las compa침칤as por a침o')
plt.legend()
plt.grid(True)
plt.show()
```

## An치lisis del gr치fico 游늼
### **Las compa침칤as Personal y Claro parecen tener una mayor presencia en el mercado en comparaci칩n con Nextel y Movistar**
#### 1. Personal muestra un crecimiento general desde 2012 hasta 2022, con algunos aumentos y disminuciones en ciertos meses y a침os. Es la compa침칤a que tiene la base de usuarios m치s grande en general.
#### 2. Claro tambi칠n muestra un crecimiento general, pero con fluctuaciones m치s notables a lo largo de los a침os. Alcanza su punto m치ximo en 2015 y luego disminuye ligeramente.
#### 3. Nextel muestra un crecimiento inicial desde 2012 hasta 2014, pero luego disminuye significativamente y parece haber perdido la mayor칤a de sus usuarios en los a침os siguientes.
#### 4. Movistar muestra un crecimiento constante desde 2012 hasta 2015, pero luego muestra una tendencia descendente hasta 2022.
#### 5. La informaci칩n de "Otros" tiene valores m치s bajos en comparaci칩n con las otras compa침칤as y no muestra un patr칩n claro de crecimiento o disminuci칩n, por eso no se logra apreciar en el gr치fico.

## N칰mero de SMS enviados
**Cantidad total de mensajes cortos de texto que se env칤an a lo largo de los a침os**

```python
sms = pd.read_csv('Sets/Movil/movil_mensajes.csv')
sms.head(40) # Nos enfocamos en el N칰mero de SMS salientes
```

```python
# Vamos a graficar los Mensajes de Textos enviados 
# y ver si disminuyen o aumentan con el paso de los a침os

# Agrupamos por a침o y calcular la suma del n칰mero de SMS salientes
sms_por_a침o = sms.groupby('A침o')['N칰mero de SMS salientes'].sum()

# Sacamos los a침os 칰nicos
a침os = sms['A침o'].unique()

# Calculamos el porcentaje de cambio en el n칰mero de SMS salientes respecto al primer a침o
porcentaje_cambio = (sms_por_a침o / sms_por_a침o.iloc[0] - 1) * 100

# Se crea el gr치fico de l칤neas
plt.figure(figsize=(8, 6))
plt.plot(a침os, porcentaje_cambio)
plt.ylabel('Porcentaje de cambio')
plt.title('Disminuci칩n del n칰mero de SMS enviados a lo largo de los a침os')
plt.xticks(rotation=45)
plt.grid(True)

# Se agrega porcentaje en el eje y
plt.gca().yaxis.set_major_formatter('{:.1f}%'.format)
plt.show()
```

```python
sms['N칰mero de SMS salientes'] = sms['N칰mero de SMS salientes'].str.replace('.', '').astype(float)
sms_por_a침o = sms.groupby('A침o')['N칰mero de SMS salientes'].sum()
porcentaje_disminucion = sms_por_a침o.pct_change() * 100
```

```python
# Disminuci칩n de los SMS salientes a침o tras a침o
porcentaje_disminucion.round(2)
```

```python
# Sumamos los porcentajes para ver cuanto es que han disminuido los SMS salientes
porcentaje_acumulado = porcentaje_disminucion.cumsum()
porcentaje_acumulado.round(2)
```

## An치lisis del gr치fico 游늼

#### 1. Hubo una disminuci칩n acelerada en el n칰mero de SMS salientes a lo largo de los a침os. 
#### Comenzando desde con un n칰mero aproximadamente 36 mil millones en el 2013, el n칰mero disminuy칩 a alrededor de 800 millones para el 2022. Es decir, desde el 2,013 hasta el 2,022 los mensajes han tenido una p칠rdida del -312.16 %
#### 2. Se puede notar una tendencia decreciente en el n칰mero de SMS salientes. Aunque hay fluctuaciones entre los a침os, la tendencia general es hacia la disminuci칩n.
#### 3. Las disminuciones m치s pronunciadas en el n칰mero de SMS salientes ocurrieron en los primeros a침os, especialmente del 2013 al 2014, donde hubo una ca칤da significativa de alrededor de 36 mil millones a 12 mil millones, o sea -51.95 %.

### **Estas observaciones sugieren que los SMS salientes han sido reemplazados en cierta medida por otras formas de comunicaci칩n, como las aplicaciones de mensajer칤a instant치nea y las redes sociales.**

## Ingresos por la prestaci칩n del servicio de telefon칤a m칩vil

```python
# Cargamos los datos
ingresos_movil = pd.read_csv('Sets/Movil/Telefonia_movil_ingresos.csv')
ingresos_movil.head()
```

```python
# Se eliminan las columnas que no vamos a usar
ingresos_movil = ingresos_movil.drop(['Trimestre','Periodo'], axis=1)
```

```python
# Eliminarmos el signo $
ingresos_movil['Ingresos (miles de $)'] = ingresos_movil['Ingresos (miles de $)'].str.replace('$', '').str.replace('.', '')
```

```python
# Se convierte a tipo de dato Float
ingresos_movil['Ingresos (miles de $)'] = ingresos_movil['Ingresos (miles de $)'].astype(float)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Agrupamos los ingresos por a침o y calcular la suma o el promedio
ingresos_por_a침o = ingresos_movil.groupby('A침o')['Ingresos (miles de $)'].sum()
# Codigo para calcular el promedio:
#ingresos_por_a침o = ingresos_movil.groupby('A침o')['Ingresos (miles de $)'].mean()

plt.figure(figsize=(10, 6))
plt.plot(ingresos_por_a침o.index, ingresos_por_a침o)
plt.xlabel('A침o')
plt.ylabel('Ingresos miles de d칩lares')
plt.title('Ingresos totales por a침o')
#plt.grid(True)

# Cambiamos el formato de los n칰meros en el eje Y
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))
plt.show()
```

Se crea una tabla para ver la suma total de ingresos a침o con a침o

```python
tabla_ingresos = ingresos_movil.pivot_table(index='A침o', values='Ingresos (miles de $)', aggfunc='sum')
#tabla_ingresos = tabla_ingresos.applymap('{:,.0f}'.format)
tabla_ingresos = tabla_ingresos.rename(columns={'Ingresos (miles de $)': 'Ingresos por a침o en miles de d칩lares'})
tabla_ingresos
```

Se crea una nueva columna que nos dice cuanto han aumentado los ingresos con respecto al a침o anterior

```python
tabla_ingresos['Porcentaje de crecimiento'] = tabla_ingresos['Ingresos por a침o en miles de d칩lares'].pct_change() * 100
tabla_ingresos
```

Se calcula el crecimiento promedio de todos los a침os

```python
# Calculamos el crecimiento promedio de porcentaje
crecimiento_promedio = tabla_ingresos['Porcentaje de crecimiento'].mean()

# Creamos la nueva columna "Crecimiento promedio"
tabla_ingresos['Crecimiento promedio'] = crecimiento_promedio
tabla_ingresos
```

Se crea una nueva columna que nos dice el porentaje de crecimiento acumulado desde el 2013 hasta el 2022

```python
tabla_ingresos['Porcentaje de crecimiento acumulado'] = tabla_ingresos['Porcentaje de crecimiento'].cumsum()
tabla_ingresos.round(2)
```

```python
tabla_ingresos.to_csv('Sets/Movil/tabla_ingresos_usar.csv', index=True)
```

#### 2. Los mayores incrementos en los ingresos se observan en los a침os 2019, 2020 y 2021, con porcentajes de crecimiento anual del 39.11%, 33.14% y 43.50% respectivamente.
#### 3. La industr칤a de telefon칤a m칩vil, es una industria fuerte y los datos demuestran que para el 2,023 seguir치 con tendendia positiva con un crecimiento promedio del 26.18 %
