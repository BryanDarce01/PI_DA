## Archivo 1
## Portabilidad numérica móvil: altas mensuales por operador
**Proporciona información sobre las preferencias de los usuarios al elegir un nuevo operador y la competencia entre los diferentes operadores en el mercado de telefonía móvil**

```python
df = pd.read_csv('Sets/Movil/Portabilidad.csv')
df.head()
```

```python
# Llenamos los valores nulos con 0
df = df.fillna(0)
```

```python
df['Año'] = df['Año'].astype('category')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertimos las columnas relevantes a tipo numérico
df['Personal'] = df['Personal'].astype(float)
df['Claro'] = df['Claro'].astype(float)
df['Nextel'] = df['Nextel'].astype(float)
df['Movistar'] = df['Movistar'].astype(float)
df['Otros'] = df['Otros'].astype(float)

# Agrupamos por año y calcular el crecimiento promedio anual de las compañías
crecimiento_personal = df.groupby('Año')['Personal'].mean().pct_change() * 100
crecimiento_claro = df.groupby('Año')['Claro'].mean().pct_change() * 100
crecimiento_nextel = df.groupby('Año')['Nextel'].mean().pct_change() * 100
crecimiento_movistar = df.groupby('Año')['Movistar'].mean().pct_change() * 100
crecimiento_otros = df.groupby('Año')['Otros'].mean().pct_change() * 100

# Sacamos los años únicos
años = df['Año'].unique()

# Creamos la gráfica de líneas para mostrar el crecimiento de las compañías
plt.figure(figsize=(8, 6))
plt.plot(años[1:], crecimiento_personal[1:], label='Personal')
plt.plot(años[1:], crecimiento_claro[1:], label='Claro')
plt.plot(años[1:], crecimiento_nextel[1:], label='Nextel')
plt.plot(años[1:], crecimiento_movistar[1:], label='Movistar')
plt.plot(años[1:], crecimiento_otros[1:], label='Otros')
plt.xlabel('Año')
plt.ylabel('Crecimiento porcentual')
plt.title('Crecimiento de las compañías por año')
plt.legend()
plt.grid(True)
plt.show()
```

## Análisis del gráfico 📑
### **Las compañías Personal y Claro parecen tener una mayor presencia en el mercado en comparación con Nextel y Movistar**
#### 1. Personal muestra un crecimiento general desde 2012 hasta 2022, con algunos aumentos y disminuciones en ciertos meses y años. Es la compañía que tiene la base de usuarios más grande en general.
#### 2. Claro también muestra un crecimiento general, pero con fluctuaciones más notables a lo largo de los años. Alcanza su punto máximo en 2015 y luego disminuye ligeramente.
#### 3. Nextel muestra un crecimiento inicial desde 2012 hasta 2014, pero luego disminuye significativamente y parece haber perdido la mayoría de sus usuarios en los años siguientes.
#### 4. Movistar muestra un crecimiento constante desde 2012 hasta 2015, pero luego muestra una tendencia descendente hasta 2022.
#### 5. La información de "Otros" tiene valores más bajos en comparación con las otras compañías y no muestra un patrón claro de crecimiento o disminución, por eso no se logra apreciar en el gráfico.

## Número de SMS enviados
**Cantidad total de mensajes cortos de texto que se envían a lo largo de los años**

```python
sms = pd.read_csv('Sets/Movil/movil_mensajes.csv')
sms.head(40) # Nos enfocamos en el Número de SMS salientes
```

```python
# Vamos a graficar los Mensajes de Textos enviados 
# y ver si disminuyen o aumentan con el paso de los años

# Agrupamos por año y calcular la suma del número de SMS salientes
sms_por_año = sms.groupby('Año')['Número de SMS salientes'].sum()

# Sacamos los años únicos
años = sms['Año'].unique()

# Calculamos el porcentaje de cambio en el número de SMS salientes respecto al primer año
porcentaje_cambio = (sms_por_año / sms_por_año.iloc[0] - 1) * 100

# Se crea el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(años, porcentaje_cambio)
plt.ylabel('Porcentaje de cambio')
plt.title('Disminución del número de SMS enviados a lo largo de los años')
plt.xticks(rotation=45)
plt.grid(True)

# Se agrega porcentaje en el eje y
plt.gca().yaxis.set_major_formatter('{:.1f}%'.format)
plt.show()
```

```python
sms['Número de SMS salientes'] = sms['Número de SMS salientes'].str.replace('.', '').astype(float)
sms_por_año = sms.groupby('Año')['Número de SMS salientes'].sum()
porcentaje_disminucion = sms_por_año.pct_change() * 100
```

```python
# Disminución de los SMS salientes año tras año
porcentaje_disminucion.round(2)
```

```python
# Sumamos los porcentajes para ver cuanto es que han disminuido los SMS salientes
porcentaje_acumulado = porcentaje_disminucion.cumsum()
porcentaje_acumulado.round(2)
```

## Análisis del gráfico 📑

#### 1. Hubo una disminución acelerada en el número de SMS salientes a lo largo de los años. 
#### Comenzando desde con un número aproximadamente 36 mil millones en el 2013, el número disminuyó a alrededor de 800 millones para el 2022. Es decir, desde el 2,013 hasta el 2,022 los mensajes han tenido una pérdida del -312.16 %
#### 2. Se puede notar una tendencia decreciente en el número de SMS salientes. Aunque hay fluctuaciones entre los años, la tendencia general es hacia la disminución.
#### 3. Las disminuciones más pronunciadas en el número de SMS salientes ocurrieron en los primeros años, especialmente del 2013 al 2014, donde hubo una caída significativa de alrededor de 36 mil millones a 12 mil millones, o sea -51.95 %.

### **Estas observaciones sugieren que los SMS salientes han sido reemplazados en cierta medida por otras formas de comunicación, como las aplicaciones de mensajería instantánea y las redes sociales.**

## Archivo 3
## Ingresos por la prestación del servicio de telefonía móvil

```python
# Cargamos los datos
ingresos_movil = pd.read_csv('Sets/Movil/Telefonia_movil_ingresos.csv')
ingresos_movil.head()
```

```python
# Se eliminan las columnas que no vamos a usar
ingresos_movil = ingresos_movil.drop(['Trimestre','Periodo'], axis=1)
```

```python
# Eliminarmos el signo $
ingresos_movil['Ingresos (miles de $)'] = ingresos_movil['Ingresos (miles de $)'].str.replace('$', '').str.replace('.', '')
```

```python
# Se convierte a tipo de dato Float
ingresos_movil['Ingresos (miles de $)'] = ingresos_movil['Ingresos (miles de $)'].astype(float)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Agrupamos los ingresos por año y calcular la suma o el promedio
ingresos_por_año = ingresos_movil.groupby('Año')['Ingresos (miles de $)'].sum()
# Codigo para calcular el promedio:
#ingresos_por_año = ingresos_movil.groupby('Año')['Ingresos (miles de $)'].mean()

plt.figure(figsize=(10, 6))
plt.plot(ingresos_por_año.index, ingresos_por_año)
plt.xlabel('Año')
plt.ylabel('Ingresos miles de dólares')
plt.title('Ingresos totales por año')
#plt.grid(True)

# Cambiamos el formato de los números en el eje Y
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))
plt.show()
```

Se crea una tabla para ver la suma total de ingresos año con año

```python
tabla_ingresos = ingresos_movil.pivot_table(index='Año', values='Ingresos (miles de $)', aggfunc='sum')
#tabla_ingresos = tabla_ingresos.applymap('{:,.0f}'.format)
tabla_ingresos = tabla_ingresos.rename(columns={'Ingresos (miles de $)': 'Ingresos por año en miles de dólares'})
tabla_ingresos
```

Se crea una nueva columna que nos dice cuanto han aumentado los ingresos con respecto al año anterior

```python
tabla_ingresos['Porcentaje de crecimiento'] = tabla_ingresos['Ingresos por año en miles de dólares'].pct_change() * 100
tabla_ingresos
```

Se calcula el crecimiento promedio de todos los años

```python
# Calculamos el crecimiento promedio de porcentaje
crecimiento_promedio = tabla_ingresos['Porcentaje de crecimiento'].mean()

# Creamos la nueva columna "Crecimiento promedio"
tabla_ingresos['Crecimiento promedio'] = crecimiento_promedio
tabla_ingresos
```

Se crea una nueva columna que nos dice el porentaje de crecimiento acumulado desde el 2013 hasta el 2022

```python
tabla_ingresos['Porcentaje de crecimiento acumulado'] = tabla_ingresos['Porcentaje de crecimiento'].cumsum()
tabla_ingresos.round(2)
```

```python
tabla_ingresos.to_csv('Sets/Movil/tabla_ingresos_usar.csv', index=True)
```

## Análisis 📑

#### 1.En general, los ingresos muestran un incremento gradual año tras año. 
#### Esto indica un crecimiento en la telefonía móvil con ingresos por año en miles de dólares.
#### 2. Los mayores incrementos en los ingresos se observan en los años 2019, 2020 y 2021, con porcentajes de crecimiento anual del 39.11%, 33.14% y 43.50% respectivamente.
#### 3. La industría de telefonía móvil, es una industria fuerte y los datos demuestran que para el 2,023 seguirá con tendendia positiva con un crecimiento promedio del 26.18 % 

### **En general, la información muestra una tendencia positiva en los ingresos, con un crecimiento significativo año tras año.**
